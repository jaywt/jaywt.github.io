
Lecture 2: Chemistry and Chemical Reactions

Oxidation state

Number of electrons lost (or gained) by an element when forming an ion
or chemical bond

Determining the oxidation state of an element in a compound:

An element bonded to itself has an oxidation state of zero (supersedes
the other rules)

Hydrogen (H) is always +1

Oxygen (O) is always -2

The sum of the oxidation states of all elements in the compound must add
up to the total charge on the compound

Examples: O2 gas, CO2, CH4, CH2O (organic matter), SeO42-

Charge Balance

Water contains positively charged (cations) and negatively charged
(anions) dissolved ions

Charge balance – Positive and negative charges must cancel each other
out

Solute charge is represented in terms of equivalents per volume (e.g.,
eq/L or meq/L for milliequivalents per liter)

eq/L = mol/L x solute charge

Unit Conversion

One of the most important skills you will ever learn is how to convert
between different units.

Examples:

What is the molarity (mol L-1) of Mg2+ in a solution prepared as 50 mg
L-1 MgCl2·6H2O?

What is the molarity (mol L-1) of Cl- in a solution prepared as 50 mg
L-1 MgCl2·6H2O?

First, what is the molecular mass (g/mol) of MgCl2-6H2O? 203.31 g/mol

(50 mg MgCl2-6H2O / L)(1 g / 1000 mg)(1 mol MgCl2-6H2O / 203.31 g
MgCl2-6H2O)(1 mol Mg / 1 mol MgCl2-6H2O) = 2.5 x 10^-4 mol/L

(50 mg MgCl2-6H2O / L)(1 g / 1000 mg)(1 mol MgCl2-6H2O / 203.31 g
MgCl2-6H2O)(2 mol Cl / 1 mol MgCl2-6H2O) = 4.9 x 10^-4 mol/L

Are these charge balanced? Yes! Cl- has one equivalent per mol (eq/mol)
and Mg2+ has two equivalents per mol.

(4.9 x 10-4 mol/L Cl-) x (1 eq/mol) = 4.9 x 10-4 eq/L

(2.5 x 10-4 mol/L Mg2+) x (2 eq/mol) = 4.9 x 10-4 eq/L

\*note that the 2.5 and 4.9 are rounded, so it’s actually like 2.45

Geochemical Models

Steps to build a model:

Define the system

Define the identity and quantity of the constituents (reactants and
products) in the system

Write a balanced equation for the chemical reaction of interest

Define the system

Closed System – composition is fixed; no mass transfer (box)

Open System – mass can enter and leave the system via fluxes (box +
arrows)

Define the constituents

Example constituents (ensure they are expressed in comparable units)

Water

Minerals

Gases (defined by their partial pressure)

Solutes (defined by their concentration; e.g., mass/volume)

Write a balanced equation for the chemical reaction of interest

Reactants Products aA + bB cC + dD

Start with an unbalanced equation of reactants and products

Make an element inventory

Adjust the numbers in front of the reactants and products until balanced
on either side of the arrow

Assuming the reaction takes place with water as a solvent, add H+, OH-
and H2O as needed to balance the equation (but only if H2O participates
in the reaction)

Double check for mass balance and charge balance!

Examples:

NaCl –\> (in water) Na+ + Cl-

MgCl2 (in water) Mg(2+) + 2 Cl(-)

Fe(3+) + H2O FeOOH (then balance the elements…) Fe(3+) + 2 H2O FeOOH + 3
H+

Chemical Reactions

Types of chemical reactions

Dissolution/precipitation

Acid-base

Adsorption/desorption

Oxidation-reduction (redox)

\*reactions can be more than one type

Dissolution reaction: AνABνB (s) νAAqA (aq) + νBBqB (aq)

A – element

νA – stoichiometric amount of A

qA – charge on the ion A

Examples:

NaCl(s) Na + Cl

CaCO3 (s) + CO2 + H2O Ca + HCO3-

Precipitation reaction: νAAqA (aq) + νBBqB (aq) AνABνB (s)

Examples:

Ca2+(aq) + SO42- (aq) + H2O gypsum

Acid-base reaction: HA + B A- + HB+

Brønsted-Lowry definition: transfer of a proton from an acid (the proton
donor) to a base (the proton acceptor)

Examples (which are the acid, base, conjugate acid, conjugate base?)

HCl + H2O Cl- + H3O+

NH3 + H2O NH4+ + OH-

CaCO3 + H2SO4 H2CO3 + Ca2+ + SO42-

Surface complexation reactions

Adsorption – the adhesion of a substance (e.g., solute) onto a surface

Desorption – the release of a substance from a surface into solution

    ≡S-O- + H+  ≡S-OH       ≡S-OH + H+  ≡S-OH2+


    ≡S-K+ + H+  ≡S-H+ + K+

Inner-sphere complex – ion forms a chemical bond with the surface at a
specific site

Outer-sphere complex – electrostatic interaction between ion and charged
surface; ion remains in hydration shell

Redox (oxidation-reduction) reactions

Transfer of electrons from e- donor (reducing agent) to e- acceptor
(oxidizing agent)

OILRIG – oxidation is loss, reduction is gain

Oxidation half reaction: A0 A+ + e- Reduction half reaction: B0 + e- B-
